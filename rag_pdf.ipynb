```python
# Importations des bibliothèques nécessaires
from langchain.chains import LLMChain
from langchain.embeddings import OpenAIEmbeddings
from langchain.prompts import PromptTemplate
from langchain.vectorstores import FAISS
import fitz
from langchain.llms import OpenAI

import os

# Pour afficher les DataFrames
import pandas as pd

# Initialisation de l'API OpenAI avec votre clé d'API


# Fonction pour extraire le texte des fichiers PDF


# Lister tous les fichiers PDF dans le répertoire 'pilly/'
source_folder = "pilly/"


# Liste des fichiers PDF dans le dossier source
pdf_files = [f for f in os.listdir(source_folder) if f.endswith('.pdf')]



# Extraire le texte de chaque fichier PDF


def extract_text_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    pdf_document.close()
    return text

documents = []

for pdf_file in pdf_files:
    pdf_path = os.path.join(source_folder, pdf_file)
    docs = extract_text_from_pdf(pdf_path)
    documents.extend(docs)

# Initialiser OpenAI embeddings
embeddings = OpenAIEmbeddings(base_url="http://localhost:1234/v1", api_key="lm-studio")

# Générer des embeddings pour chaque document


# Initialiser FAISS vector store
vectorstore = FAISS.from_documents(documents, embeddings)



print("Les embeddings ont été générés et stockés dans FAISS.")

```

```python
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.document_loaders import PyMuPDFLoader
from langchain.vectorstores.faiss import FAISS
from langchain.embeddings import OpenAIEmbeddings
import pickle
import os

print("Loading data...")
pdf_folder_path = "pilly/pdfs/"
pdf_files = os.listdir(pdf_folder_path)
print(pdf_files)

# Load multiple files
loaders = [PyMuPDFLoader(os.path.join(pdf_folder_path, fn)) for fn in pdf_files]

all_documents = []

for loader in loaders:
    print("Loading raw document..." + loader.file_path)
    raw_documents = loader.load()

    print("Splitting text...")
    text_splitter = CharacterTextSplitter(
        separator="\n\n",
        chunk_size=800,
        chunk_overlap=100,
        length_function=len,
    )
    documents = text_splitter.split_documents(raw_documents)

    # Extract page content if it's a Document object
    for doc in documents:
        if hasattr(doc, 'page_content'):
            all_documents.append(doc.page_content)
        elif isinstance(doc, str):
            all_documents.append(doc)
        else:
            raise ValueError(f"Unexpected document type: {type(doc)}")

print("Creating vectorstore...")
# Ensure all_documents is a list of strings
if not all(isinstance(doc, str) for doc in all_documents):
    raise ValueError("All documents should be strings")

embeddings = OpenAIEmbeddings(base_url="http://localhost:1234/v1", api_key="lm-studio")
vectorstore = FAISS.from_documents(all_documents, embeddings)

with open("vectorstore.pkl", "wb") as f:
    pickle.dump(vectorstore, f)

```
