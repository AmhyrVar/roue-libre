Cleaning du prompt pour avoir un json valid

```python
import fitz  # PyMuPDF
from openai import OpenAI
import os
import json
import tiktoken


# Initialisation de l'API OpenAI
openai_api_key = ''
openai = OpenAI(api_key=openai_api_key)

# Dossier contenant les PDF
pdf_folder = "pilly/"

# Liste des fichiers PDF dans le dossier
pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]

def num_tokens(text):
    encoding = tiktoken.encoding_for_model("gpt-4o")
    return len(encoding.encode(text))

# Fonction pour extraire le texte d'un PDF
def extract_text_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    pdf_document.close()
    return text

# Few-shot exemples pour le prompt
few_shot_examples = """
{
    "qtype": "symptômes",
    "Question": "Quels sont les symptômes de la brucellose aiguë septicémique ?",
    "Answer": "Les symptômes de la brucellose aiguë septicémique incluent une fièvre ondulante sudoro-algique, des douleurs musculaires et articulaires, des courbatures, des sueurs abondantes (nocturnes), et une asthénie. L'examen clinique peut révéler une hépatomégalie et une splénomégalie dans 25 à 40 % des cas, ainsi que quelques adénopathies périphériques et des râles bronchiques."
},
{
    "qtype": "diagnostic",
    "Question": "Comment diagnostique-t-on la brucellose ?",
    "Answer": "Le diagnostic de la brucellose repose sur des éléments d'orientation comme le contexte épidémiologique et la leuconeutropénie habituelle associée à un syndrome inflammatoire modéré. Les examens spécifiques incluent les hémocultures, la recherche du germe dans des liquides biologiques ou des prélèvements tissulaires, la sérologie avec différents tests (séroagglutination de Wright, fixation du complément, card-test, IFI, ELISA), et la technique d’amplification génique (PCR)."
},
"""

# Fonction pour générer des entrées à partir du texte extrait
def generate_entries_from_pdf_text(pdf_text):
    # Split the PDF text into two parts
    halfway_point = len(pdf_text) // 2
    pdf_text_part1 = pdf_text[:halfway_point]
    pdf_text_part2 = pdf_text[halfway_point:]
    
    few_shot_examples = """
    # Add your few-shot examples here
    """
    
    def get_response_from_openai(pdf_text_part):
        prompt = f"""
        {few_shot_examples}
        En te basant sur le PDF que je t'ai donné et les few-shot examples, génère-moi un json valide avec des entrées question réponse et qtype sur les traitements, symptômes, diagnostic, épidémiologie, etc. Ne te base que sur le PDF que je t'ai donné, génère-moi 30 entrées.
        Ta réponse ne doit contenir que ce qui devrait être entre des "[]" car elle devra être parsée par json.loads()
        PDF Content:
        {pdf_text_part}
        """
        response = openai.chat.completions.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=4096,
            n=1,
            stop=None,
            temperature=0
        )
        
        return response.choices[0].message.content
    
    # Get responses for both parts of the PDF
    response_part1 = get_response_from_openai(pdf_text_part1)
    response_part2 = get_response_from_openai(pdf_text_part2)
    
    # Calculate token usage for both parts
    total_tokens_used = num_tokens(response_part1) + num_tokens(response_part2)
   
    
    # Merge the generated JSON entries
    generated_entries_part1 = json.loads(response_part1)
    generated_entries_part2 = json.loads(response_part2)
    merged_entries = generated_entries_part1 + generated_entries_part2
    
       
   
    return json.dumps(merged_entries, ensure_ascii=False, indent=2)


# Processus principal pour traiter chaque PDF
for pdf_file in pdf_files:
    try:
        pdf_path = os.path.join(pdf_folder, pdf_file)
        pdf_text = extract_text_from_pdf(pdf_path)
        entries = generate_entries_from_pdf_text(pdf_text)
        
        json_output = f"{pdf_folder}{os.path.splitext(pdf_file)[0]}_entries.json"
        
        # Conversion en format JSON
        with open(json_output, 'w', encoding='utf-8') as f:
            json.dump(json.loads(entries), f, ensure_ascii=False, indent=4)

        print(f"Entries for {pdf_file} generated and saved to {json_output}")
    except Exception as e:
        print(f"Error processing {pdf_file}: {e}")
        continue

print("Traitement terminé.")
```

```python
import os
import shutil

# Dossier source et destination
source_folder = "pilly/"
destination_folder = "pilly/completed/"

# Créer le dossier de destination s'il n'existe pas
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Liste des fichiers PDF dans le dossier source
pdf_files = [f for f in os.listdir(source_folder) if f.endswith('.pdf')]

# Fonction pour déplacer les fichiers
def move_files(source, destination, filename):
    src_file = os.path.join(source, filename)
    dest_file = os.path.join(destination, filename)
    shutil.move(src_file, dest_file)
    print(f"{filename} moved to {destination}")

# Processus principal pour déplacer chaque PDF et son fichier JSON correspondant
for pdf_file in pdf_files:
    # Générer le nom du fichier JSON correspondant
    json_file = os.path.splitext(pdf_file)[0] + "_entries.json"
    
    # Vérifier si le fichier JSON correspondant existe
    if os.path.exists(os.path.join(source_folder, json_file)):
        print(f"Déplacement {pdf_file}, et {json_file}.")
        # Déplacer le fichier PDF
        move_files(source_folder, destination_folder, pdf_file)
        
        # Déplacer le fichier JSON correspondant
        move_files(source_folder, destination_folder, json_file)
    else:
        print(f"No matching JSON for {pdf_file}, skipping.")

print("Le déplacement des paires de fichiers PDF et JSON est terminé.")
```

```python
import os
import json

# Folder containing the JSON files
json_folder = "pilly/"

# List of JSON files in the folder
json_files = [f for f in os.listdir(json_folder) if f.endswith('.json')]

# Empty list to store the data from each JSON file
data = []


# Read each JSON file and append its data to the list
for json_file in json_files:
    json_path = os.path.join(json_folder, json_file)
    with open(json_path, 'r', encoding='utf-8') as f:
        json_data = json.load(f)
        data.extend(json_data)



# Path of the merged JSON file
merged_json_path = "pilly_dataset_full.json"

# Write the merged data to the JSON file
with open(merged_json_path, 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print(f"The JSON files in {json_folder} have been merged into {merged_json_path}.")
```
