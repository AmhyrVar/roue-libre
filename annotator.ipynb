```python
import json

# Specify the path to the JSON file
file_path = 'maladies.json'

# Open the JSON file and load its contents
with open(file_path, 'r') as file:
    data = json.load(file)

# Now you can work with the data from the JSON file
# For example, you can access specific values using keys
data
```

```python
import re

# Iterate over the elements of the data list
for i in range(len(data)):
    # Remove content between parentheses using regular expressions
    data[i] = re.sub(r'\(.*?\)', '', data[i])

# Print the modified data list
data
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
#from langchain.schema import SimpleJsonOutputParser # StrOutputParser

#from langchain_core.output_parsers.json import SimpleJsonOutputParser

from langchain_core.output_parsers import JsonOutputParser

template = PromptTemplate.from_template("""
Vous êtes un agent qui écrit des courtes phrases médicales.
Ecrivez des courtes phrases médicales en français qu'un médecin écrirait dans un rapport médical.
La phrase médicale écrite doit essentiellement comporter la pathologie telle que donnée textuellement par l'utilisateur cette expression ne doit jamais changer.
Ne mentionnez dans la phrase aucun mot ou expression qui pourrait faire référence à une autre pathologie ou maladie.
N'écrivez jamais d'expression introductif du genre ceci est votre phrase.
Veillez a n'écrire qu'une seule phrase.
                                        
Génère moi une phrase comportant cette pathologie : {pathologie}
""")


llm_chain = (template | llm )

def générateur(pathologie):
    response = llm_chain.invoke({"pathologie": pathologie})
    return response
```

```python
data_limited = data[:10]
```

```python
result_dict = {}
for element in data_limited:
    result_dict[element] = générateur(element)
    print(element)
    print(result_dict[element])
    print("-------------------------------------------------------------------------")
```

```python
for k,v in result_dict.items():
    print(k)
    print(v)
    print("**************")
```

```python
import json

# Specify the file path to save the JSON file
output_file_path = 'result_dict.json'

# Save the result_dict variable as a JSON file
with open(output_file_path, 'w') as file:
    json.dump(result_dict, file)
```
